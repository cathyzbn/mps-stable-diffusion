```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
filename = 'thread_results.csv'
df = pd.read_csv(filename)
df.head()
```

```python
def visualize():
    fig, ax = plt.subplots()
    sns.barplot(x='num_workers', y='throughput', hue='compute_type', data=df, ax=ax)
    plt.title(f'Throughput vs. Number of Threads / Processes')
    plt.xlabel('Number of Threads / Processes')
    plt.ylabel('Throughput (requests/s)')
    plt.legend(title='Configuration', bbox_to_anchor=(1.05, 1), loc='upper left')
    plt.show()
```

```python
visualize()
```
