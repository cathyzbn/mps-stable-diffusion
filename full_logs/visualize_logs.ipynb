```python
import pandas as pd
import matplotlib.pyplot as plt
import glob
import os
```

```python
path = os.getcwd()
all_files = glob.glob(path + "/*.csv")
all_files.sort()
```

```python
# Iterate over all files and plot the queue times
for file in all_files:
    # Read the CSV file
    df = pd.read_csv(file)
    df['time'] = df['start_time'] - df['start_time'][0]
    
    # Extract the filename without the path and extension for the legend
    filename = os.path.basename(file).replace('.csv', '')
    
    # Plot the queue times
    plt.plot(df['time'], df['queue_time'], label=filename)

# Set plot title and labels
plt.title('Queue Times of Requests over time')
plt.xlabel('Time')
plt.ylabel('Queue Time')

# Show legend
plt.legend(title='Filename')

# Display the plot
plt.show()
```
